```python
import pandas as pd
```

```python
columnsToDrop = ['CASE','SERIAL','REF', 'QUESTNNR', 'MODE', 'STARTED', 'TIME_SUM', 'MAILSENT', 'LASTDATA', 'FINISHED', 'Q_VIEWER', 'LASTPAGE', 'MAXPAGE', 'MISSING', 'MISSREL', 'TIME_RSI']
for i in range(1, 9):
    columnsToDrop.append('TIME00' + i.__str__())
```

```python
susColumns = []
for i in range(1, 11):
    suffix = '0'
    if i == 10:
        suffix = ''
    susColumns.append('SU01_' + suffix + i.__str__())
```

```python
df: pd.DataFrame = pd.read_excel('data/data_you-sum.xlsx')
df.drop([0], inplace=True)
df.drop(columns=columnsToDrop, inplace=True)
df.reset_index(drop=True)
```

```python
video_columns = []
for video_idx in range(1,4):
    for question_idx in range(1, 4):
        for subq_idx in range(1,5):
            video_columns.append(f"V{video_idx}0{question_idx}_0{subq_idx}")
```

```python
# multiple choice questions, 1= not selected (NO), 2=selected (YES)
correct_answers = {
    'V101_01': 2,
    'V101_02': 1,
    'V101_03': 1,
    'V101_04': 2,

    'V102_01': 1,
    'V102_02': 1,
    'V102_03': 2,
    'V102_04': 1,

    'V103_01': 1,
    'V103_02': 2,
    'V103_03': 1,
    'V103_04': 1,


    'V201_01': 1,
    'V201_02': 2,
    'V201_03': 1,
    'V201_04': 1,

    'V202_01': 1,
    'V202_02': 1,
    'V202_03': 1,
    'V202_04': 2,

    'V203_01': 1,
    'V203_02': 2,
    'V203_03': 1,
    'V203_04': 1,


    'V301_01': 1,
    'V301_02': 1,
    'V301_03': 2,
    'V301_04': 1,

    'V302_01': 2,
    'V302_02': 1,
    'V302_03': 1,
    'V302_04': 1,

    'V303_01': 1,
    'V303_02': 1,
    'V303_03': 1,
    'V303_04': 2
}
```

```python
def calc_perc_for_entry(df_entry):
    score = 0
    for column in video_columns:
        if df_entry[column] == correct_answers[column]:
            score = score + 1
    return score / len(video_columns)
```

```python
df['perc'] = df.apply(lambda row: calc_perc_for_entry(row), axis=1)
```

```python
df = df[df["perc"] >= 0.75]
```
